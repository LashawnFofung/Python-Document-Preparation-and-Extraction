# Work with JSON Data

**Data:** *sample.json* file


```python
#Code will load the JSON fiile into Python using the json library
import json


# Load JSON file
with open('sample.json', 'r', encoding='utf-8') as file:
    data = json.load(file)


print(f"Data loaded successfully! Found {len(data)} customers.")
```

```python
# Extract all order IDs and their corresponding total amounts from JSON file
order_totals = []


for customer in data:
    for order in customer['orders']:
        order_totals.append({
            'order_id': order['order_id'],
            'total': order['total']
        })


# Display results
for order in order_totals:
    print(f"Order ID: {order['order_id']} | Total: ${order['total']:.2f}")
```

```python
#Normalizes the JSON structure using Pandas to create a flat DataFrame showing customer_id, order_id, product_id, quantity, and total

import pandas as pd


# Normalize the nested JSON structure
normalized_records = []


for customer in data:
    customer_id = customer['customer_id']

    for order in customer['orders']:
        order_id = order['order_id']
        order_total = order['total']

        for item in order['items']:
            record = {
                'customer_id': customer_id,
                'order_id': order_id,
                'product': item['product'],
                'quantity': item['quantity'],
                'total': order_total
            }
            normalized_records.append(record)


# Create DataFrame
df = pd.DataFrame(normalized_records)
print(df)
```

```python
# Convert DataFrame back to JSON
df.to_json("processed_sample.json", orient="records", indent=4)
```
