```python
# upload from Google Drive mount
#Extract text from the first page

import pdfplumber

file_path = '/content/drive/My Drive/sample_title_report.pdf'

# Extract text from the first page
try:
    with pdfplumber.open(file_path) as pdf:
        # pdf.pages[0] selects the first page
        page = pdf.pages[0]
        print(f"--- Text from the first page of {file_path.split('/')[-1]} ---")
        print(page.extract_text())

except FileNotFoundError:
    print(f"ERROR: File not found. Double-check the path: {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

Extract Tables

```python
# file_path from Google Drive
# Extract tables from the first page

import pdfplumber

try:
    with pdfplumber.open(file_path) as pdf:
        # pdf.pages[12] selects the 13th page (since indexing starts at 0)
        page = pdf.pages[12]
        tables = page.extract_tables()

        print(f"--- Tables extracted from Page 13 of {file_path.split('/')[-1]} ---")

        # Loop through each table found on the page
        for i, table in enumerate(tables):
            print(f"\n[Table {i+1}]:")

            # Loop through each row in the current table
            for row in table:
                print(row)

except FileNotFoundError:
    print(f"ERROR: File not found. Double-check the path: {file_path}")
except IndexError:
    print("ERROR: Page 13 (index 12) does not exist in this PDF. Check the total page count.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
#load data from Google Drive
#file_path = '/content/drive/My Drive/sample_title_report.pdf'
# Extract text from the first page

import fitz #PyMuPDF


try:
    # Load the PDF
    doc = fitz.open(file_path)

    # Extract text from the first page (index 0)
    page = doc[0]
    text = page.get_text()

    print(f"--- Extracted Text from Page 1 of '{file_path.split('/')[-1]}' ---")
    print(text)

    # Close the document to release resources
    doc.close()

except FileNotFoundError:
    print(f"ERROR: File not found. Double-check the path: {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

Extract text As A Structured Dictionary

```python
#load data from Google Drive
#file_path = '/content/drive/My Drive/sample_title_report.pdf'


import fitz  # PyMuPDF

page_number = 0  # Set the page number you want to extract from (0 is the first page)

try:
    # Load the PDF document
    doc = fitz.open(file_path)

    # Get the specific page object
    page = doc[page_number]

    # Extract text as a structured dictionary ("dict" output)
    text_data = page.get_text("dict")

    print(f"--- Structured Text Data from Page {page_number + 1} of '{file_path.split('/')[-1]}' ---")

    for block in text_data.get("blocks", []):  # Loop through all blocks in the page
        bbox = block["bbox"]  # Get bounding box of the block

        # Check if 'lines' exists in the block
        if "lines" in block:
            for line in block["lines"]:
                for span in line.get("spans", []):  # Ensure spans exist
                    text = span["text"]
                    print(f"Text: {text}, Bounding Box: {bbox}")

    # Close the document to release resources
    doc.close()

except FileNotFoundError:
    print(f"ERROR: File not found. Double-check the path: {file_path}")
except IndexError:
    print(f"ERROR: Page {page_number + 1} does not exist in this PDF.")
except Exception as e:
    print(f"An error occurred: {e}")
```

Render The First Page As An Image

```python
#load data from Google Drive
#file_path = '/content/drive/My Drive/sample_title_report.pdf'

# Render the first page as an image

import fitz  # PyMuPDF

page_index = 0    # 0 for the first page

try:
    # Load the PDF document
    doc = fitz.open(file_path)

    # Get the specific page object (This defines 'page')
    page = doc[page_index]

    # Render the page as a Pixmap object
    pix = page.get_pixmap()

    # Save the pixmap to a file in the Colab local file system
    output_filename = "page1_render.png" # Using a unique name
    pix.save(output_filename)

    # Close the document to release resources
    doc.close()

    print(f"Successfully loaded '{file_path.split('/')[-1]}' and saved page {page_index + 1} as '{output_filename}'.")
    print("Check the 'Files' tab on the left to see the image and download it.")

except FileNotFoundError:
    print(f"ERROR: File not found. Double-check the path: {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```
